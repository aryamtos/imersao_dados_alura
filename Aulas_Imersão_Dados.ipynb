Os valores negativos são onversamente correlacionados, quando um sobe o outro desce, quando temos valores muito próximos de 1 temos valores proporcionamlmente correlacionados, ou seja, quando um sobe o outro sobe também.
```python
corr = dados.loc[:,'g-0':'g-50'].corr()
```

```python
import numpy as np
```

```python
corr = dados.loc[:,'g-0':'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap,  center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

A impressão que temos é que existe uma correlação baixa entre os valores. Importante lembrar que correlação não implica em causalidade -> spurious correlations

O c é a viabilidade celular quantas células sobreviveram a exposição do composto

```python
corr_celular = dados.loc[:,'c-0':'c-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap,  center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

O tipo celular deve está influenciando na sobrevivência se um tipo celular sobrevive o outro provavelmente vai sobreviver também.Alguns genes vão ter a correlação alta para todos os tipos celulares

# Aula 4 - Merge e Análise de dados

```python
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true'
resultados = pd.read_csv(url_dados)
resultados
```

Aqui temos os mecanismo s de ação das drogas. A interação bioquímica do composto que é a droga com o alvo pode ser uma célula uma proteína até mesmo de um patógeno que pode está causando uma doença. Esse mecanismo de ação traz duas informações o alvo e ação que ele tem em sobre o alvo. O alvo aí é a alpha_reductase, portanto esse mecnismo de ação está reduzindo a alpha reductase. O agonista é o que está potencializando aquele receptor, ativando e o inibidor é o que está inibindo, cabe destcar também que o antagonista é o que bloqueia.

```python
resultados['acat_inhibitor'].unique()
```

Quando um determinado mecanismo de ação é ativado ele é colocado como 1, caso contrário 0. 

```python
contagem_moa = resultados.drop('id',axis=1).sum().sort_values(ascending=False)
contagem_moa
```

Vamos verificar se a linha 1 foi ativada em algum momento

```python
 resultados.drop('id',axis=1).sum(axis=1)
```

```python
 resultados.head()
```

Algum controle ativou algum mecanismo de ação? O esperado é que não

```python
resultados['n_moa'] = resultados.drop('id',axis=1).sum(axis=1)
```

```python
resultados['ativo_moa'] = (resultados['n_moa'] != 0)
resultados.head()
```

Merge de dados

```python
dados_combinados = pd.merge(dados, resultados[['id','n_moa','ativo_moa']],on='id')
dados_combinados.head()
```

Com controle não teve nenhum mecanismo de ação

```python
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
```

```python
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

Os cinco compostos que mais aparecem e desses cinco, apenas um não ativa 

```python
composto_principal = dados_combinados['composto'].value_counts().index[:5]
#composto_principal
plt.figure(figsize=(12,8))
sns.boxplot(data=dados_combinados.query('composto in @composto_principal'), y ='g-0',x='composto', hue='ativo_moa')
```

<h1>Árvore de Decisão</h1>

```python
from sklearn.tree import DecisionTreeClassifier
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino,x_teste, y_treino,y_teste = train_test_split(x,y, test_size=0.2,stratify =y,random_state=376)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino,y_treino)
modelo_arvore.score(x_teste,y_teste)
```

```python
from sklearn import tree
fig, ax = plt.subplots(figsize=(15,10), facecolor='k')
tree.plot_tree(modelo_arvore,
                ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names=x_treino.columns,
               class_names=['Não Ativado', 'Ativado'])
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino,x_teste, y_treino,y_teste = train_test_split(x,y, test_size=0.2,stratify =y,random_state=376)

teste=[]
treino=[]

for i in range(1,15):
  modelo_arvore = DecisionTreeClassifier(max_depth=1)
  modelo_arvore.fit(x_treino,y_treino)
  teste.append(modelo_arvore.score(x_teste,y_teste))
  treino.append(modelo_arvore.score(x_treino,y_treino))
```
